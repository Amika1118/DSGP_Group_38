```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Load the dataset
file_path = '../data/raw/Vegetable_prices_weekly.csv'
df = pd.read_csv(file_path)

# Display the first few rows
df.head()
```

```python
# Shape of the dataset
print(f"Dataset shape: {df.shape}")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

```python
# Column names and their data types
df.info()
```

```python
# Display column names in a readable list
print("Column names:")
for col in df.columns:
    print(f"  - {col}")
```

```python
# First few rows (already displayed above) and last few rows
df.tail()
```

```python
# Summary statistics for all numerical columns
df.describe(include=[np.number]).T
```

```python
# List of categorical columns (excluding time-related if needed)
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns:", cat_cols)
```

```python
# Value counts for 'vegetable' column
print("Vegetable counts:")
print(df['vegetable'].value_counts())
```

```python
# Value counts for each actual_class column
for col in [c for c in cat_cols if 'actual_class' in c]:
    print(f"\n{col}:")
    print(df[col].value_counts(dropna=False))
```

```python
# Count of missing values per column
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Percentage': missing_pct})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
print("Columns with missing values:")
print(missing_df)
```

```python
# Visualize missing values with a heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# If many missing, we can also plot bar chart
missing_df.plot(kind='bar', y='Percentage', legend=False, figsize=(12,5))
plt.title('Percentage of Missing Values by Column')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# List of target vegetables
target_veggies = ['Carrot', 'Cabbage', 'Tomatoes', 'Brinjals', 'Pumpkin', 'Bitter Gourd']

# Create a copy for analysis (no modification to original)
df_target = df[df['vegetable'].isin(target_veggies)].copy()
print(f"Records for target vegetables: {len(df_target)}")
print(f"Original dataset records: {len(df)}")
```

```python
# Count per vegetable in the target subset
df_target['vegetable'].value_counts()
```

```python
# Boxplot of price by vegetable
plt.figure(figsize=(12,6))
sns.boxplot(data=df_target, x='vegetable', y='price')
plt.title('Price Distribution by Vegetable')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histograms for each vegetable
g = sns.FacetGrid(df_target, col='vegetable', col_wrap=3, sharex=False, sharey=False)
g.map(plt.hist, 'price', bins=30, edgecolor='black')
g.set_axis_labels('Price (LKR)', 'Frequency')
g.fig.suptitle('Price Histograms for Each Vegetable', y=1.02)
plt.show()
```

```python
# Summary statistics of price per vegetable
df_target.groupby('vegetable')['price'].describe()
```

```python
# Convert year and week to a date-like format for plotting
# We'll create a 'week_year' column for sorting
df_target['week_year'] = df_target['year'].astype(str) + '-' + df_target['week']

# Alternatively, create a continuous week number (optional)
# For simplicity, we'll use year-week as category but sort properly.
# Better: create a date column assuming week starts from week 1 of each year.
# We'll use a helper: convert week to week number and then to ordinal.
# But for quick visualization, we can plot with year-week on x-axis (may be crowded).
```

```python
# For time series, we can plot average price per week for each vegetable
# First, ensure week is sorted chronologically
# Create a sort key: year * 100 + week_number (extract numeric from 'W1')
df_target['week_num'] = df_target['week'].str.extract('(\d+)').astype(int)
df_target['year_week'] = df_target['year'] * 100 + df_target['week_num']
df_target = df_target.sort_values('year_week')

# Plot price over time for each vegetable
plt.figure(figsize=(14,8))
for veg in target_veggies:
    veg_data = df_target[df_target['vegetable'] == veg]
    plt.plot(veg_data['year_week'], veg_data['price'], label=veg, alpha=0.7)
plt.xlabel('Time (Year-Week)')
plt.ylabel('Price (LKR)')
plt.title('Price Trends Over Time by Vegetable')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# List all actual_class columns
actual_class_cols = [col for col in df.columns if 'actual_class' in col]
print("Climate class columns:", actual_class_cols)
```

```python
# For each region, plot the distribution of classes
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(actual_class_cols):
    if i < len(axes):
        df[col].value_counts(dropna=False).plot(kind='bar', ax=axes[i], title=col.replace('_actual_class',''))
        axes[i].set_xlabel('Class')
        axes[i].set_ylabel('Count')

# Hide unused subplots
for j in range(i+1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle('Distribution of Actual Climate Classes by Region', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# For a given region (e.g., Badulla), plot price distribution for each class
# We'll do this for each vegetable separately

region = 'Badulla'
class_col = f'{region}_actual_class'

plt.figure(figsize=(14,8))
for i, veg in enumerate(target_veggies, 1):
    plt.subplot(2,3,i)
    veg_data = df_target[df_target['vegetable'] == veg]
    sns.boxplot(data=veg_data, x=class_col, y='price')
    plt.title(f'{veg} - {region} Class')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# List precipitation columns
precip_cols = [col for col in df.columns if 'precipitation' in col]
print("Precipitation columns:", precip_cols)
```

```python
# Summary statistics for precipitation
df[precip_cols].describe()
```

```python
# Distribution of precipitation for each region (histograms)
df[precip_cols].hist(bins=30, figsize=(15,10), edgecolor='black')
plt.suptitle('Precipitation Distributions by Region', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Time series of precipitation for each region (weekly)
# We'll use the same year_week sort key as before
df['week_num'] = df['week'].str.extract('(\d+)').astype(int)
df['year_week'] = df['year'] * 100 + df['week_num']
df = df.sort_values('year_week')

plt.figure(figsize=(14,8))
for col in precip_cols:
    plt.plot(df['year_week'], df[col], label=col.replace('_precipitation',''), alpha=0.7)
plt.xlabel('Time (Year-Week)')
plt.ylabel('Precipitation')
plt.title('Weekly Precipitation Trends by Region')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Compute correlation between price and each region's precipitation for a specific vegetable
veg = 'Carrot'
veg_data = df_target[df_target['vegetable'] == veg]
corr_precip = veg_data[precip_cols + ['price']].corr()['price'].drop('price')
print(f"Correlation of {veg} price with precipitation:\n", corr_precip)
```

```python
# Visualize as heatmap for all vegetables
# Create a DataFrame of correlations between price and precip for each veg
corr_matrix = pd.DataFrame(index=target_veggies, columns=precip_cols)
for veg in target_veggies:
    veg_data = df_target[df_target['vegetable'] == veg]
    corr = veg_data[precip_cols + ['price']].corr()['price'].drop('price')
    corr_matrix.loc[veg] = corr.values

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix.astype(float), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation between Price and Precipitation by Vegetable')
plt.show()
```

```python
# List probability columns
prob_drought_cols = [col for col in df.columns if 'prob_drought' in col]
prob_flood_cols = [col for col in df.columns if 'prob_flood_risk' in col]
prob_normal_cols = [col for col in df.columns if 'prob_normal' in col]

print("Drought probability columns:", prob_drought_cols)
print("Flood probability columns:", prob_flood_cols)
print("Normal probability columns:", prob_normal_cols)
```

```python
# Summary statistics for each set
print("Drought probabilities:")
df[prob_drought_cols].describe().T
```

```python
print("Flood probabilities:")
df[prob_flood_cols].describe().T
```

```python
print("Normal probabilities:")
df[prob_normal_cols].describe().T
```

```python
# For a specific region, check if drought + flood + normal  1
region = 'Badulla'
drought_col = f'{region}_prob_drought'
flood_col = f'{region}_prob_flood_risk'
normal_col = f'{region}_prob_normal'

sum_prob = df[[drought_col, flood_col, normal_col]].sum(axis=1)
print(f"Sum of probabilities for {region}: min={sum_prob.min()}, max={sum_prob.max()}, mean={sum_prob.mean()}")
# Ideally they should be close to 1.
```

```python
# Plot histograms for drought probabilities across regions
df[prob_drought_cols].hist(bins=30, figsize=(15,10), edgecolor='black')
plt.suptitle('Drought Probability Distributions by Region', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Correlation between price and economic indicators per vegetable
eco_cols = ['USD_LKR_avg', 'RateChange_avg_%']
corr_eco = pd.DataFrame(index=target_veggies, columns=eco_cols)
for veg in target_veggies:
    veg_data = df_target[df_target['vegetable'] == veg]
    corr = veg_data[eco_cols + ['price']].corr()['price'].drop('price')
    corr_eco.loc[veg] = corr.values

print("Correlation between Price and Economic Indicators:")
corr_eco
```

```python
# Select numerical columns (exclude year, week if needed, but week is categorical)
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# Remove 'year' and 'week_num' if they are in there (they are numeric but not for correlation)
num_cols = [c for c in num_cols if c not in ['year', 'week_num', 'year_week']]
print("Numerical columns for correlation:", num_cols)
```

```python
# Compute correlation matrix
corr_matrix_full = df[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(20,16))
sns.heatmap(corr_matrix_full, annot=False, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix of All Numerical Variables')
plt.tight_layout()
plt.show()
```

```python
# Focus on correlations with 'price'
price_corr = corr_matrix_full['price'].sort_values(ascending=False)
print("Top correlations with price:")
print(price_corr.head(20))
```

```python
# Average price per week-of-year across all years
df_target['week_of_year'] = df_target['week_num']  # week_num is 1-52

# Compute average price per week for each vegetable
weekly_avg = df_target.groupby(['vegetable', 'week_of_year'])['price'].mean().reset_index()

# Plot
plt.figure(figsize=(14,8))
for veg in target_veggies:
    veg_data = weekly_avg[weekly_avg['vegetable'] == veg]
    plt.plot(veg_data['week_of_year'], veg_data['price'], label=veg, marker='o', markersize=3)
plt.xlabel('Week of Year')
plt.ylabel('Average Price (LKR)')
plt.title('Seasonal Pattern: Average Price by Week of Year')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Average price per year for each vegetable
yearly_avg = df_target.groupby(['vegetable', 'year'])['price'].mean().reset_index()

plt.figure(figsize=(14,8))
for veg in target_veggies:
    veg_data = yearly_avg[yearly_avg['vegetable'] == veg]
    plt.plot(veg_data['year'], veg_data['price'], label=veg, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Price (LKR)')
plt.title('Yearly Average Price Trends')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import os

# Load the dataset
input_path = "../data/preprocessed/vegetable_prices_clean.csv"
df = pd.read_csv(input_path)

print("Dataset loaded successfully.")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Columns to drop
columns_to_drop = ['year', 'week', 'year_week']
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

print("Unnecessary columns removed.")
print(f"Remaining columns: {list(df.columns)}")
df.head()
```

```python
# Mapping for actual_class columns
class_mapping = {
    'normal': 0,
    'flood_risk': 1,
    'drought': -1
}

# List of actual_class columns
class_columns = [
    'Badulla_actual_class',
    'Hambantota_actual_class',
    'Jaffna_actual_class',
    'Kurunegala_actual_class',
    'Matale_actual_class',
    'Nuwara_Eliya_actual_class',
    'Ratnapura_actual_class'
]

# Apply mapping
for col in class_columns:
    if col in df.columns:
        df[col] = df[col].map(class_mapping)
    else:
        print(f"Warning: Column {col} not found in dataset.")

print("Actual_class columns encoded.")
# Check unique values after encoding
for col in class_columns:
    if col in df.columns:
        print(f"{col}: {df[col].unique()}")
df.head()
```

```python
# Mapping for vegetable names
veg_mapping = {
    'Bitter Gourd': 1,
    'Brinjals': 2,
    'Cabbage': 3,
    'Carrot': 4,
    'Pumpkin': 5,
    'Tomatoes': 6
}

# Encode the vegetable column
if 'vegetable' in df.columns:
    df['vegetable'] = df['vegetable'].map(veg_mapping)
    print("Vegetable column encoded.")
    print(f"Unique values in vegetable: {df['vegetable'].unique()}")
else:
    print("Warning: 'vegetable' column not found.")
df.head()
```

```python
# Check data types and basic info
print("Data types after encoding:")
print(df.dtypes)

# Check for any missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Display a random sample to confirm encoding
df.sample(10)
```

```python
# Define output path
output_dir = "../data/feature_engineered"
output_path = os.path.join(output_dir, "vegetable_prices_fe.csv")

# Create directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Save to CSV
df.to_csv(output_path, index=False)
print(f"Featureengineered dataset saved to: {output_path}")
```

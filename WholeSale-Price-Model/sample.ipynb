```python
import pandas as pd
from pathlib import Path

# Full path to your Excel file
excel_file = Path(r"C:\Users\Arosha IIT\OneDrive - Robert Gordon University\Desktop\DSGP\WholeSale-Price-Model\data\raw\Amika.xlsx")

# Check if the file exists
if not excel_file.exists():
    raise FileNotFoundError(f"File not found: {excel_file.resolve()}")

# Read the Excel file
df = pd.read_excel(excel_file)

# Save as CSV in the same folder
csv_file = excel_file.with_suffix(".csv")  # keeps the same name, changes extension to .csv
df.to_csv(csv_file, index=False)

print(f"Excel file has been successfully converted to CSV: {csv_file.resolve()}")
```

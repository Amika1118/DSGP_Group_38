```python
import pandas as pd
# Load historical weekly retail + fuel data
historical_df = pd.read_csv("/content/retail_fuel_weekly_preprocessed.csv")

# Load future wholesale prediction data
future_df = pd.read_csv("/content/realistic_weekly_predictions_2026_2040.csv")
```

```python
#Inspect and standardize date columns
print(future_df.columns)
```

```python
future_df['Date'] = pd.to_datetime(future_df['Date'])
future_df = future_df.rename(columns={'Date': 'date'})
```

```python
future_df['date'] = pd.to_datetime(future_df['date'])
historical_df['Start_Date'] = pd.to_datetime(historical_df['Start_Date'])
```

```python
# Convert future dates to week start (Monday)
future_df['Start_Date'] = future_df['date'] - pd.to_timedelta(
    future_df['date'].dt.weekday, unit='D'
)
```

```python
#Rename columns for consistency
future_df = future_df.rename(columns={
    'vegetable': 'Variety',
    'predicted_price': 'Predicted_Wholesale_Price'
})
```

```python
#Add Data_Type flags
historical_df['Data_Type'] = 'Historical'
future_df['Data_Type'] = 'Forecast'
```

```python
#Rename columns properly
future_df = future_df.rename(columns={
    'Vegetable_Name': 'Variety',
    'Predicted_Price_Rs_kg': 'Predicted_Wholesale_Price'
})
```

```python
#Confirm columns
print(future_df.columns)
```

```python
future_df = future_df[['Start_Date', 'Variety',
                       'Predicted_Wholesale_Price', 'Data_Type']]
```

```python
#Add empty column for alignment
historical_df['Predicted_Wholesale_Price'] = pd.NA
```

```python
#Combine historical + future datasets
final_df = pd.concat(
    [historical_df, future_df],
    ignore_index=True
)
```

```python
#Sort final dataset by date & variety
final_df = final_df.sort_values(
    by=['Variety', 'Start_Date']
).reset_index(drop=True)
```
